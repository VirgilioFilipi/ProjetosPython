```python
import numpy as np
import pandas as pd
from faker import Faker
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import random
import os 
import cassino_DB
fake = Faker(['pt_BR'])
import cassino_query 
```

```python
n_player = 5000
n_betting = 100000
n_game = 25

db_file = 'Banco/cassino.db'


db_manager = cassino_DB.CassinoDB(n_player, n_betting, n_game)
query_manager = cassino_query.CassinoQuery(db_file)
```

```python
players_ages = query_manager.fetch_all_ages()

print(players_ages)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(players_ages, bins=20, kde=True)
plt.title('Distribuio das Idades dos Jogadores')
plt.xlabel('Idade')
plt.ylabel('Frequncia')
plt.show()

```
